```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset[['TV']], dataset[['Sales']], test_size = 0.3,random_state=0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# to make predictions with a trained machine learning model on a test dataset and then display those predictions.
res= model.predict(X_test)
print(res)
```

```python
#coefficient
model.coef_
```

```python
#intercept
model.intercept_
```

```python
# creating a line plot of the predicted values stored in the res variable using Matplotlib.
plt.plot(res)
```

```python
plt.figure(figsize=[7,5])
plt.scatter(y_train,X_train)
plt.xlabel("Original values")
plt.ylabel("Predicted values")
plt.title("Original values vs Predicted values")
plt.show()
```
